Parte 1: Programacion Estructurada


```python
%%bash
cat > TiendaEstructurada.java <<'EOF'
import java.util.Scanner;

public class TiendaEstructurada {
    public static void main(String [] args) {
        Scanner sc = new Scanner(System.in);

        System.out.print("Ingrese el nombre del producto: ");
        String nombre = sc.nextLine();

        System.out.print("Ingrese el precio unitario: ");
        double precio = sc.nextDouble();

        System.out.print("Ingrese la cantidad: ");
        int cantidad = sc.nextInt();

        double total = calcularTotal(precio , cantidad);

        mostrarResumen(nombre , precio , cantidad , total);
    }

    public static double calcularTotal(double precio , int cantidad) {
        return precio * cantidad;
    }

    public static void mostrarResumen(String nombre , double precio , int cantidad , double total) {
        System.out.println("\n=== RESUMEN DE PRODUCTO ===");
        System.out.println("Producto: " + nombre);
        System.out.println("Precio unitario: $" + precio);
        System.out.println("Cantidad: " + cantidad);
        System.out.println("Total: $" + total);
    }
}
EOF

# Compilar
javac TiendaEstructurada.java

# Ejecutar con entradas de ejemplo usando un heredoc (puedes modificar las líneas siguiente para probar otros valores)
java TiendaEstructurada <<EOF
Manzana
0.75
10
EOF

```

Parte 2: Programaci´on Orientada a Objetos

```python
%%bash
cat > Producto.java <<'EOF'
class Producto {
    private String nombre;
    private double precio;
    private int cantidad;

    public Producto(String nombre , double precio , int cantidad) {
        this.nombre = nombre;
        this.precio = precio;
        this.cantidad = cantidad;
    }

    public double calcularTotal() {
        return precio * cantidad;
    }

    public void mostrarInfo() {
        System.out.println("\nProducto: " + nombre);
        System.out.println("Precio unitario: $" + precio);
        System.out.println("Cantidad: " + cantidad);
        System.out.println("Total: $" + calcularTotal());
    }
}
EOF

cat > TiendaApp.java <<'EOF'
import java.util.Scanner;

public class TiendaApp {
    public static void main(String [] args) {
        Scanner sc = new Scanner(System.in);

        System.out.println("=== REGISTRO DE PRODUCTOS ===");
        Producto[] productos = new Producto[2];
        double totalGeneral = 0;

        for (int i = 0; i < 2; i++) {
            System.out.println("\nProducto #" + (i + 1));
            System.out.print("Nombre: ");
            String nombre = sc.nextLine();
            System.out.print("Precio: ");
            double precio = sc.nextDouble();
            System.out.print("Cantidad: ");
            int cantidad = sc.nextInt();
            sc.nextLine(); // limpiar buffer

            productos[i] = new Producto(nombre , precio , cantidad);
        }

        System.out.println("\n=== RESUMEN DE INVENTARIO ===");
        for (Producto p : productos) {
            p.mostrarInfo();
            totalGeneral += p.calcularTotal();
        }

        System.out.println("\nTOTAL GENERAL DEL INVENTARIO: $" + totalGeneral);
    }
}
EOF

# Compilar ambas clases
javac *.java

# Ejecutar con entradas de ejemplo (modifica las líneas para probar otros productos)
java TiendaApp <<EOF
Pan
1.20
5
Leche
0.95
8
EOF
```

Preguntas de reflexion

1.¿Que ventajas observas en la version orientada a objetos?

Encapsulamiento: cada Producto guarda su estado (nombre, precio, cantidad) y se encarga de calcular su total. Eso reduce código repetido y errores.
Reutilización: puedes usar la clase Producto en otros programas sin copiar/pegar lógica.

Abstracción: el programa principal (app) trata con objetos Producto y ya no necesita saber los detalles internos.

Mantenibilidad: si cambias cómo se calcula el total (descuentos, impuestos), actualizas solo Producto y todo el sistema recibe la mejora.

Escalabilidad: añadir nuevas funciones (inventario, reportes, persistencia) es más natural: agregas clases (Tienda, Factura, Repositorio) sin romper lo existente.

4. ¿Podrıas agregar una clase Tienda que contenga los productos y calcule el total?

Sí — abajo tienes la implementación completa con Tienda que usa ArrayList<Producto> y métodos:
agregarProducto(Producto p)
calcularTotalInventario()
mostrarInventario()

```python
%%bash
cat > Producto.java <<'EOF'
public class Producto {
    private String nombre;
    private double precio;
    private int cantidad;

    public Producto(String nombre, double precio, int cantidad) {
        this.nombre = nombre;
        this.precio = precio;
        this.cantidad = cantidad;
    }

    public String getNombre() {
        return nombre;
    }

    public double getPrecio() {
        return precio;
    }

    public int getCantidad() {
        return cantidad;
    }

    public double calcularTotal() {
        return precio * cantidad;
    }

    public void mostrarInfo() {
        System.out.printf("Producto: %s | Precio unitario: $%.2f | Cantidad: %d | Total: $%.2f%n",
                          nombre, precio, cantidad, calcularTotal());
    }
}
EOF
```

```python
%%bash
cat > Tienda.java <<'EOF'
import java.util.ArrayList;
import java.util.List;

public class Tienda {
    private List<Producto> productos;

    public Tienda() {
        productos = new ArrayList<>();
    }

    public void agregarProducto(Producto p) {
        productos.add(p);
    }

    public double calcularTotalInventario() {
        double total = 0.0;
        for (Producto p : productos) {
            total += p.calcularTotal();
        }
        return total;
    }

    public void mostrarInventario() {
        System.out.println("\\n=== INVENTARIO DE LA TIENDA ===");
        if (productos.isEmpty()) {
            System.out.println("No hay productos registrados.");
            return;
        }
        for (Producto p : productos) {
            p.mostrarInfo();
        }
        System.out.printf("%nTOTAL GENERAL DEL INVENTARIO: $%.2f%n", calcularTotalInventario());
    }

    public int cantidadProductos() {
        return productos.size();
    }
}
EOF
```

```python
%%bash
cat > TiendaApp.java <<'EOF'
import java.util.Scanner;

public class TiendaApp {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        Tienda tienda = new Tienda();

        System.out.println("=== SISTEMA DE REGISTRO DE PRODUCTOS (Tienda) ===");

        int n;
        if (args.length >= 1) {
            // Si se pasa n por argumento use ese valor
            n = Integer.parseInt(args[0]);
            System.out.println("Registrando " + n + " productos (desde argumentos).");
        } else {
            System.out.print("¿Cuántos productos desea registrar?: ");
            while (!sc.hasNextInt()) {
                System.out.print("Por favor ingrese un número entero: ");
                sc.next();
            }
            n = sc.nextInt();
            sc.nextLine(); // limpiar buffer
        }

        for (int i = 0; i < n; i++) {
            System.out.println("\\nProducto #" + (i + 1));

            System.out.print("Nombre: ");
            String nombre = sc.nextLine().trim();
            while (nombre.isEmpty()) {
                System.out.print("Nombre no puede estar vacío. Ingréselo de nuevo: ");
                nombre = sc.nextLine().trim();
            }

            double precio;
            System.out.print("Precio (ej: 1.25): ");
            while (true) {
                if (sc.hasNextDouble()) {
                    precio = sc.nextDouble();
                    sc.nextLine();
                    if (precio >= 0) break;
                    System.out.print("El precio debe ser >= 0. Ingréselo de nuevo: ");
                } else {
                    System.out.print("Valor inválido. Ingréselo de nuevo: ");
                    sc.next();
                }
            }

            int cantidad;
            System.out.print("Cantidad: ");
            while (true) {
                if (sc.hasNextInt()) {
                    cantidad = sc.nextInt();
                    sc.nextLine();
                    if (cantidad >= 0) break;
                    System.out.print("La cantidad debe ser >= 0. Ingrésela de nuevo: ");
                } else {
                    System.out.print("Valor inválido. Ingréselo de nuevo: ");
                    sc.next();
                }
            }

            Producto p = new Producto(nombre, precio, cantidad);
            tienda.agregarProducto(p);
        }

        // Mostrar inventario y total
        tienda.mostrarInventario();

        System.out.println("\\nFin del programa.");
    }
}
EOF
```
